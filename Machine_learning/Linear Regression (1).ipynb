# Linear Regression

Steps:

  - Show chart with all data
  - Calculate line equation
  - Calcultate Loss function
  - Trainning

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def displayData(path, xLabel, yLabel):
    X, Y = np.loadtxt(path, skiprows=1, unpack=True)
    plt.xlabel(xLabel, fontsize=20)
    plt.ylabel(yLabel, fontsize=20)
    plt.axis([0, 50, 0, 50])
    plt.plot(X, Y, "bo")
    plt.show()
```

```python
displayData("/Users/ys322r/Documents/Dev/Machine Learning/how-machine-learning-works/01/demos/code/steaks.txt", "Reservation", "Steaks")
```

Calculate line equetion for each reservation

```python
def predict(X, w, b):
    return X * w + b
```

```python
def loss(X, Y, w, b):
    predictions = predict(X, w, b)
    return np.average((predictions - Y) ** 2)
```

```python
def train(X, Y, iterations, lr):
    w = b = 0 # random start point
    for i in range(iterations):
        current_loss = loss(X, Y, w, b)
        print("Iteration %4d => Loss %.6f" % (i, current_loss))
        if (loss(X, Y, w - lr, b) < current_loss):
            w -= lr
        elif loss(X, Y, w + lr, b) < current_loss:
            w += lr
        elif loss(X, Y, w, b - lr) < current_loss:
            b -= lr
        elif loss(X, Y, w, b + lr) < current_loss:
            b += lr
        else:
            return w, b
    return w, b
```

```python
def main(X, Y, iterations, lr) :
    w,b = train(X, Y, iterations, lr)
    plt.xlabel("Reservations", fontsize=20)
    plt.ylabel("Steaks", fontsize=20)
    plt.axis([0, 50, 0, 50])
    plt.plot(X, Y, "bo")

    # Plot the line:
    plt.plot([0, 50], [b, predict(50, w, b)], color="g")

    # Visualize the diagram:
    plt.show()
```

```python
main(X, Y, 10000, 0.001)
```
