```python
import matplotlib.pyplot as plt
x = [5, 8 , 10]
y = [12, 16, 6]

plt.plot(x,y)

plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('ggplot')

x1 = [5, 8 , 10]
y1 = [12, 16, 6]

x2 = [6, 9 , 11]
y2 = [6, 15, 7]



plt.plot(x1, y1, 'g', label='line one', linewidth=5)
plt.plot(x2, y2, 'c', label='line two', linewidth=5)

plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.legend()
plt.grid(True, color='k')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt



x = np.random.uniform(low=0, high=50, size=(100,))
y = np.random.uniform(low=0, high=50, size=(100,))

plt.xlabel("X axis", fontsize=20)
plt.ylabel("Y axis", fontsize=20)
plt.axis([0, 50, 0, 50])
plt.plot(x, y, "bo")
plt.show()
```
